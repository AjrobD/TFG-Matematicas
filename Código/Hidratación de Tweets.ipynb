```python
client = tweepy.Client(bearer_token=bearer_token,
                       consumer_key=consumer_key,
                       consumer_secret=consumer_secret,
                       access_token=access_token,
                       access_token_secret=access_token_secret,
                       wait_on_rate_limit=True
                       )
```

# C - Hidrataci贸n

```python
with open('FIFA_2014_sentiment_dataset.txt') as f:
    lines = f.readlines()
```

Introducimos cada linea del txt en un diccinario, con key el id y value una lista con el sentiment

```python
cont = 0
dict_tweets = dict()
for line in lines:
  cont += 1
  ids, sentiment = line.replace('=','').split()
  dict_tweets[int(ids)] = [sentiment]

print(cont)
print(len(dict_tweets))
```

Es posible que haya ids repetidos, por tanto, se comprueba.

```python
lista_ids = list(dict_tweets.keys())
print(len(lista_ids))
```

Recuperamos los tweets con su correspondiente id en tandas de 100 ids y lo introducimos en el diccionario

```python
batch_size = 100 
lista_ids_batches = [lista_ids[i:i + batch_size] for i in range(0, len(lista_ids), batch_size)]

for list_tweet_ids in lista_ids_batches:
  list_tweet = client.get_tweets(list_tweet_ids)

  for tweet in list_tweet[0]:
    tweet_id = int(tweet.data['id'])
    if len(dict_tweets[tweet_id]) < 2:
      dict_tweets[tweet_id].append(tweet.data['text'])
```

Eliminamos aquellos ids cuyos tweets no hemos sido capaces de recuperar

```python
cont = 0
copy_dict_tweets = dict(dict_tweets)
for item in copy_dict_tweets.items():
  if (len(item[1])) < 2:
    cont += 1
    print(item)
    dict_tweets.pop(item[0])
print(cont)
```

Comprobamos que las 10 primeras y 煤ltimas entradas del diccionario son correctas

```python
print(list(dict_tweets.items())[:10])
print(list(dict_tweets.items())[len(dict_tweets)-10:])
```

```python
with open("FIFA_2014_sentiment_dataset_processed.json", "w") as outfile:
    json.dump(dict_tweets, outfile)
```
